# Split Dataset Into train,test

```python
from sklearn.model_selection import train_test_split
x_data = dummy_data.drop(['Production', 'Yield'], axis=1)
target_feature = dummy_data['Production']


# Splitting dataset - 25% into test and 75% into train

x_train, x_test, y_train, y_test = train_test_split(x_data,target_feature,test_size=0.25, random_state=0)
```

# 1. Linear Regression


```python
# importing linear regression module from sklearn library
from sklearn.linear_model import LinearRegression
L_model = LinearRegression()
```

```python
# L_model.fit(x_train,y_train)

import pickle
 
# Save the trained model as a pickle string.
saved_model = pickle.dumps(L_model)
 
# Load the pickled model
knn_from_pickle = pickle.loads(saved_model)
 
# Use the loaded pickled model to make predictions
knn_from_pickle.predict(x_test)
```

```python
#Predicting the Test Results
L_prediction = L_model.predict(x_test)
L_prediction
```

R & R2 score

```python
from sklearn.metrics import r2_score
r2_l= r2_score(y_test, L_prediction)
print('R2 score:', r2_l)
```

```python
# Calculating Adj. R2 score: 
Adjr2_l = 1 - (1-r2_l)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
print("Adj. R-Squared : {}".format(Adjr2_l))
```

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator = L_model, X = x_train, y=y_train, cv = 10)
# Mean Accuracy and SD of 10 fold results
a_l = (accuracies.mean()*100)
print("Accuracy : {:.2f}%".format (accuracies.mean()*100))
```

```python
plt.scatter(y_test, L_prediction)
plt.xlabel('Actual Values')
plt.ylabel('Predicted values')
plt.title('Linear Regression Model')
```

# 3.Stacked LSTM model

```python
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error as mse
```

```python
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(np.array(dummy_data))
scaled_data

training_size=int(len(scaled_data)*0.70)
test_size=len(scaled_data)-training_size
train_data,test_data=scaled_data[0:training_size,:],scaled_data[training_size:len(scaled_data),:1]
training_size, test_size

import numpy
def create_dataset(dataset, time_step=1):
    data_x, data_y = [], []
    for i in range(len(dataset)-time_step-1):
        a= dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
        data_x.append(a)
        data_y.append(dataset[i + time_step, 0])
    return numpy.array(data_x), numpy.array(data_y)


# reshape into X=t,t+1,t+2,t+3 and Y=t+4
time_step = 100
x_train, y_train = create_dataset(train_data, time_step)
x_test, y_test = create_dataset(test_data, time_step)

# Reshaping intput data to be [sample, time step, features] which is required for LSTM
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)
x_test  = x_test.reshape(x_test.shape[0], x_test.shape[1], 1 )

x_train.shape, x_test.shape
```

```python
#Building Model
LSTM_model=Sequential()
LSTM_model.add(LSTM(50, return_sequences=True,input_shape=(100,1)))
LSTM_model.add(LSTM(50, return_sequences=True))
LSTM_model.add(LSTM(50))
LSTM_model.add(Dense(1,activation='softmax'))
```

```python
#Compiling the model
LSTM_model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
```

```python
LSTM_model.summary()
```

```python
#Plotting loss, it shows that loss has decreased significantly and model trained well
loss=LSTM_model.history.history['loss']
plt.plot(loss)
```

```python
# Lets do the prediction and performance 
train_predict = LSTM_model.predict(x_train)
test_predict  = LSTM_model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_lstm = r2_score(y_test, test_predict)
print('R2 Score :',r2_lstm )
```

```python
# Adjust R2 score
Adjr2_lstm= 1 - (1-r2_lstm)*(len(y_test)-1)/(len(y_test)-x_test.shape[1]-1)
print("Adjust R-Squared : {}".format(Adjr2_lstm))
```
